```python
import matplotlib.pyplot as plt
import numpy as np

# Valores de n
n = np.arange(1, 21)

# Funções de complexidade
O_1 = np.ones_like(n)
O_log_n = np.log2(n)
O_n = n
O_n_log_n = n * np.log2(n)
O_n_squared = n ** 2
O_2_power_n = 2 ** n

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(n, O_1, label="O(1) - Constante", marker='o')
plt.plot(n, O_log_n, label="O(log n) - Logarítmica", marker='o')
plt.plot(n, O_n, label="O(n) - Linear", marker='o')
plt.plot(n, O_n_log_n, label="O(n log n) - Linearítmica", marker='o')
plt.plot(n, O_n_squared, label="O(n²) - Quadrática", marker='o')
plt.plot(n, O_2_power_n, label="O(2^n) - Exponencial", marker='o')

# Detalhes do gráfico
plt.yscale('log')  # Escala logarítmica para facilitar a visualização
plt.xlabel("Tamanho da Entrada (n)")
plt.ylabel("Complexidade de Tempo")
plt.title("Comparação de Ordens de Crescimento de Complexidade de Tempo")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Valores de n
n = np.arange(1, 21)

# Funções de complexidade para o pior caso (Big-O), melhor caso (Omega) e caso médio (Theta)
O_1 = np.ones_like(n)
O_log_n = np.log2(n)
O_n = n
O_n_log_n = n * np.log2(n)
O_n_squared = n ** 2
O_2_power_n = 2 ** n

# Pior caso (Big-O)
O_worst_case = [O_n_squared, O_2_power_n]

# Melhor caso (Omega)
Omega_best_case = [O_1, O_n]

# Caso médio (Theta)
Theta_average_case = [O_n_log_n, O_log_n]

# Plotando o gráfico
plt.figure(figsize=(10, 6))

# Pior caso (Big-O)
plt.plot(n, O_n_squared, label="O(n²) - Pior Caso (Big-O)", color='r', linestyle='--')
plt.plot(n, O_2_power_n, label="O(2^n) - Pior Caso (Big-O)", color='r', linestyle='--')

# Melhor caso (Omega)
plt.plot(n, O_1, label="Ω(1) - Melhor Caso (Omega)", color='g', linestyle=':')
plt.plot(n, O_n, label="Ω(n) - Melhor Caso (Omega)", color='g', linestyle=':')

# Caso Médio (Theta)
plt.plot(n, O_n_log_n, label="Θ(n log n) - Caso Médio (Theta)", color='b', linestyle='-')
plt.plot(n, O_log_n, label="Θ(log n) - Caso Médio (Theta)", color='b', linestyle='-')

# Detalhes do gráfico
plt.yscale('log')  # Escala logarítmica para facilitar a visualização
plt.xlabel("Tamanho da Entrada (n)")
plt.ylabel("Complexidade de Tempo")
plt.title("Comparação de Ordens de Crescimento (Big-O, Omega, Theta)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo valores de profundidade (d)
d = np.arange(1, 10)

# Complexidade temporal teórica para cada variante
b = 2  # Fator de ramificação médio para simplificação
A_star_standard = b ** d
A_star_admissible = d ** 1.5  # Heurística admissível simplificada
IDA_star = b ** d
Memory_bounded_A_star = b ** (d - 1)  # Simulação com uso de memória limitada
Dynamic_weighted_A_star = b ** (d * 0.9)  # Menor expansão devido ao peso

# Plotando o gráfico
plt.figure(figsize=(12, 6))
plt.plot(d, A_star_standard, label="A* Padrão - O(b^d)", marker='o', color='blue')
plt.plot(d, A_star_admissible, label="A* Admissível - O(d^1.5)", marker='o', color='orange')
plt.plot(d, IDA_star, label="IDA* - O(b^d)", marker='o', color='green')
plt.plot(d, Memory_bounded_A_star, label="Memory-Bounded A* - O(b^(d-1))", marker='o', color='red')
plt.plot(d, Dynamic_weighted_A_star, label="Dynamic Weighted A* - O(b^(0.9*d))", marker='o', color='purple')

# Configurações do gráfico
plt.yscale('log')  # Escala logarítmica para facilitar a visualização
plt.xlabel("Profundidade do Caminho (d)")
plt.ylabel("Nós Expandidos (Complexidade Temporal)")
plt.title("Comparação da Complexidade Temporal das Variantes do Algoritmo A*")
plt.legend()
plt.grid(True)
plt.show()
```
